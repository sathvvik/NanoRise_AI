```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
# Load the dataset
df = pd.read_csv("feedback_sentiment.csv")

# Check the first few rows of the dataset
df.head()
```

```python
# Convert feedback text to lowercase
df['feedback'] = df['feedback'].str.lower()

# Remove special characters (optional)
df['feedback'] = df['feedback'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)

df.head()
```

```python
sia = SentimentIntensityAnalyzer()

# Get sentiment score (positive, negative, neutral)
df['sentiment_score'] = df['feedback'].apply(lambda x: sia.polarity_scores(x)['compound'])

# Display the dataframe with sentiment scores
df.head()
```

```python
X = df['feedback']  # Feedback text
y = df['sentiment']  # Actual sentiment labels
```

```python
# Initialize CountVectorizer to convert text into numerical data
vectorizer = CountVectorizer(stop_words='english')

# Transform the text data into numerical form
X_vec = vectorizer.fit_transform(X)

# Check the shape of the transformed data
X_vec.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_vec, y, test_size=0.2, random_state=42)
```

```python
# Train a Naive Bayes classifier
model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
# Make predictions on the test data
y_pred = model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# Ask the user for their feedback input
new_feedback = input("Enter your feedback: ")

# Convert the input feedback into features using CountVectorizer
new_feedback_vec = vectorizer.transform([new_feedback])

# Predict the sentiment of the feedback using the model
prediction = model.predict(new_feedback_vec)

# Display the result: 'Positive' if predicted as 1, 'Negative' if predicted as 0
if prediction[0] == 1:
    print("Feedback: Positive")
else:
    print("Feedback: Negative")

```
